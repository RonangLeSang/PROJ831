```python
import json
```

```python
json_file_path = "data/extracted_information/crimes_category.json"

with open(json_file_path, 'r') as jsonfile:
    data_category = json.load(jsonfile)

print(data_category)

json_file_path = "data/extracted_information/charge_description_by_degree.json"

with open(json_file_path, 'r') as jsonfile:
    data_degree = json.load(jsonfile)

print(data_degree)
```

```python
def find_key_by_value(dictionary, target_value):
    for key, values in dictionary.items():
        for value in values:
            if value == target_value:
                return key
    return None

degrees_with_category = {}

for degree in data_degree.keys():
    degrees_with_category[degree] = []
    for crime in data_degree[degree]:
        category = find_key_by_value(data_category, crime)
        if not category in degrees_with_category[degree]:
            degrees_with_category[degree].append(category)

json_file_path = "data/extracted_information/crimes_degrees_with_category.json"

with open(json_file_path, 'w') as jsonfile:
    json.dump(degrees_with_category, jsonfile)

degrees_with_category
```
